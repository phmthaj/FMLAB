### 🧩 Phân tích mô tả các thuộc tính chính

---

#### (A) Cột `Age`
| Phân vị | Giá trị |
|:--------|:--------|
| 25% | 35 |
| 50% (Median) | 45 |
| 75% | 55 |

→ Tuổi khách hàng tập trung quanh **35–55 tuổi**, khá hợp lý.  
Không có ngoại lệ (`min = 23`, `max = 67`).  
**Không cần xử lý.**

---

#### (B) Cột `Experience`
| Thống kê | Giá trị |
|:----------|:---------|
| Min | -3 *(bất hợp lý)* |
| Mean | 20.10 |
| Max | 43 |

→ Min âm, không hợp lý.  
Sẽ **điều chỉnh hoặc loại bỏ** các hàng có giá trị `< 0`.

---

#### (C) Cột `Income`
| Phân vị | Giá trị (nghìn USD/năm) |
|:--------|:-------------------------|
| 25% | 39 |
| 50% | 64 |
| 75% | 98 |
| Max | 224 |

→ Phân phối **lệch phải mạnh (right-skewed)**.  
Phần lớn khách hàng thu nhập thấp/trung bình, một số ít thu nhập rất cao (outlier).  
Sẽ **chuẩn hóa bằng Min–Max Normalization hoặc StandardScaler.**

---

#### (D) Cột `CCAvg` (Credit Card Average Spending)
| Phân vị | Giá trị (nghìn USD/tháng) |
|:--------|:---------------------------|
| 25% | 0.7 |
| 50% | 1.5 |
| 75% | 2.5 |
| Max | 10 |

→ **Lệch phải mạnh** — phần lớn khách hàng tiêu < 3k USD/tháng,  
nhưng có vài người tiêu đến 10k → sẽ **tìm cách chuẩn hóa.**

---

#### (E) Cột `Education`
| Giá trị | Nghĩa |
|:--------|:------|
| 1 | Undergraduate (Đại học) |
| 2 | Graduate (Thạc sĩ) |
| 3 | Advanced/Professional (Tiến sĩ, chuyên gia cao cấp) |

→ Trung bình: **1.88** → phần lớn khách hàng ở mức đại học hoặc thạc sĩ.

---

#### (F) Cột `Mortgage`
| Phân vị | Giá trị (nghìn USD) |
|:--------|:---------------------|
| 25% | 0 |
| 50% | 0 |
| 75% | 101 |
| Max | 635 |

→ Hơn một nửa khách **không có khoản thế chấp** (median = 0).  
Một số ít có **mortgage rất lớn** (max = 635).  
→ Biến **lệch phải** → cần **clip hoặc log-transform.**

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
import matplotlib.pyplot as plt

X = df2.drop(columns=['ID', 'Personal Loan'])
y = df2['Personal Loan']

print("Trước SMOTE:", Counter(y))

smote = SMOTE(sampling_strategy=0.5, random_state=42, k_neighbors=5)
X_res, y_res = smote.fit_resample(X, y)

print("Sau SMOTE:", Counter(y_res))

plt.figure(figsize=(5,4))
plt.bar(['Label 0 (No Loan)', 'Label 1 (Loan)'], Counter(y_res).values(), color=['skyblue','salmon'])
plt.title(' Phân bố nhãn sau khi áp dụng SMOTE (tỷ lệ 0.5)')
plt.ylabel('Số mẫu')
plt.show()
```

Sử dụng SMOTE để làm data bớt chênh lệch, tạo thêm số mẫu 1 bằng SMOTE

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2,random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

knn = KNeighborsClassifier()

# --- Lưới tham số cần tìm ---
param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11, 13, 15],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}

grid_search = GridSearchCV(
    estimator=knn,
    param_grid=param_grid,
    scoring='f1',       
    cv=5,               
    n_jobs=-1,          
    verbose=2
)
```

```python
grid_search.fit(X_train, y_train)

print("Best Parameters:", grid_search.best_params_)
print("Best F1-score (CV):", round(grid_search.best_score_, 4))
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

results = pd.DataFrame(grid_search.cv_results_)
pivot = results.pivot_table(
    values='mean_test_score',
    index='param_n_neighbors',
    columns='param_metric'
)

plt.figure(figsize=(7,5))
sns.heatmap(pivot, annot=True, fmt=".3f", cmap="YlGnBu")
plt.title("🔍 F1-score trung bình theo số láng giềng (k) và metric")
plt.ylabel("Số láng giềng (k)")
plt.xlabel("Metric")
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix, classification_report

best_knn = KNeighborsClassifier(
    n_neighbors=3,
    weights='distance',
    metric='manhattan'
)

best_knn.fit(X_train, y_train)
```

```python
y_pred = best_knn.predict(X_test)

acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Hiệu suất trên tập Test:")
print(f"Accuracy : {acc:.4f}")
print(f"Precision: {prec:.4f}")
print(f"Recall   : {rec:.4f}")
print(f"F1-score : {f1:.4f}")

print("\nBáo cáo chi tiết:")
print(classification_report(y_test, y_pred, digits=3))

print("\nMa trận nhầm lẫn:")
print(confusion_matrix(y_test, y_pred))
```
